## With all the dataset

### Data

#### Load data

```python
df, labels, names = pd.read_pickle("../data/dataset_COVID_RAW.pkl")
df = pd.DataFrame(df)
labels = pd.Series(labels, name = 'label')
names = pd.Series(names, name = 'names')
df = pd.concat((df, labels, names), axis = 1)
```

```python
X_set = df.drop(columns = ['label', "names"]).values
Y_set = df.label.values
groups = df.names.values
```

```python
folds = list(LeaveOneGroupOut().split(X_set, Y_set, groups = groups))
```

#### Create sets

```python
train_dataset, validation_dataset, test_dataset = setsCreation("../train_settings/training_settings_cov_raw.pckl", X_set, Y_set, folds)
```

#### Visualize some data

```python
cov = df[df['label']==0]
cov = cov.drop(columns = ['label', "names"]).values[0]
covNeg = df[df['label']==1]
covNeg = covNeg.drop(columns = ['label', "names"]).values[3]
ctrl = df[df['label']==2]
ctrl = ctrl.drop(columns = ['label', "names"]).values[0]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,15))

ax1.plot(cov, 'r')
ax1.set_xlabel('Raman shift (cm-1)')
ax1.set_ylabel('Intensity (a.u.)')
ax1.set_title('Covid positive',fontsize= 15)

ax2.plot(covNeg, 'm')
ax2.set_xlabel('Raman shift (cm-1)')
ax2.set_ylabel('Intensity (a.u.)')
ax2.set_title('Covid negative',fontsize= 15)

ax3.plot(ctrl, 'c')
ax3.set_xlabel('Raman shift (cm-1)')
ax3.set_ylabel('Intensity (a.u.)')
ax3.set_title('Control',fontsize= 15)

fig.tight_layout()

plt.show()
```

### Test
